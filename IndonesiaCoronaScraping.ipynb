```python
# made based on hobbies --- Fahri---
import requests
from bs4 import BeautifulSoup
import re
```

```python
r = requests.get('https://www.covid19.go.id/')
soup = BeautifulSoup(r.text, 'html.parser')
print(soup)
```

```python
page = soup.find("div",{"class":"fusion-layout-column fusion_builder_column fusion_builder_column_1_4 fusion-builder-column-5 fusion-one-fourth 1_4"})
data = page.get_text()
```

```python
print(data)
```

```python
p = '[\d]+[.,\d]+|[\d]*[.][\d]+|[\d]+'
dataNumber = []
if re.search(p, str(data)) is not None:
    for catch in re.finditer(p, str(data)):
        print(catch[0])
        dataNumber.append(catch[0])

print(dataNumber)
print('Positif : %s , Sembuh : %s , Meninggal : %s' % (dataNumber[0],dataNumber[1],dataNumber[2]))
```
